```python
data['team1'].value_counts() + data['team2'].value_counts()
# returns the total no. of matches played by each team 
```

```python
(data['team1'].value_counts() + data['team2'].value_counts()).sort_values()
# returns the total no. of matches played by each team in ascending order 
```

```python
(data['team1'].value_counts() + data['team2'].value_counts()).sort_values(ascending=False)
# returns the total no. of matches played by each team in descending order 

# here the changes are temporary , if we pass the argument inplace=True in sort_values then the changes are permanent 
```

```python
data.sort_values('city',ascending=True)  

# city column is sorted in ascending order 
```

```python
data.sort_values(['city','date'],ascending=[True,False]) 

# sorts the city column in ascending order 
# sorts the date column in descending order 
```

```python
data.drop_duplicates(subset=['city'])

# returns the unique cities , by default it keeps the first element and drops other elements , to keep the last value we use the argument keep='last'
```

```python
data.drop_duplicates('season',keep='last')

# final match of every season
```

```python
data.drop_duplicates('season',keep='last')[['season','winner']].sort_values('season')

# winners of every season 
```
