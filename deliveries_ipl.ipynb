```python
def run_scored(batsman_name) :
    max_run=delivery[delivery['batsman']==batsman_name]
    return max_run.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1)

# creating a function which returns the team name for which the batsman scored maximum runs
```

```python
run_scored('MS Dhoni') 
```

```python
# isin() fn.
# Find the most destructive death over batsman in the history of ipl
# Strike rate=(Number of runs/Number of balls)*100
# Batsman min 200 balls in over 16-20 
```

```python
death_over=delivery[delivery['over']>15]
```

```python
death_over

# considers only those rows where the over is between 16 and 20 
```

```python
death_over.groupby('batsman')['batsman_runs'].count() 

# No. of balls played by batsman in death over 
```

```python
death_over.groupby('batsman')['batsman_runs'].count()>=200 

# considering only those batsman who have played more than 200 balls in death over
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
x=all_batsman>=200
all_batsman[x]

# No. of balls played by the batsman in death over 
```

```python
all_batsman[x].shape 

# total 44 batsman who have played minimum 200 balls in death over
```

```python
batsman_list=all_batsman[x].index.tolist() 
# Creates a list consisting of the death over batsman 
```

```python
batsman_list
```

```python
 # Runs scored by all these 43 batsman 
# Balls played by these 43 batsman 

final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum() 

# Total runs scored by the batsman 
```

```python
balls=final.groupby('batsman')['batsman_runs'].count() 

# Total balls played by the batsman 
```

```python
sr=(runs/balls)*100
```

```python
sr
```

```python
sr.sort_values(ascending=False) 

# Top deathover batsman with maximum strike rate
```
